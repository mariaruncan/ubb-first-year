```python
t = var('t')
x = function('x')(t)
y = function('y')(t)
eqd1 = diff(x,t) == 2*x - 5*y
eqd2 = diff(y,t) == 5*x + 2*y
syst = [eqd1, eqd2]
C1, C2 = var('C1, C2')
show(desolve_system(syst, [x, y], ics = [0, C1, -C2]))
```
