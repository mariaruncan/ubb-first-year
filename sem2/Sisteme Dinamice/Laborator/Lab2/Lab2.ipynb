```python
#1a
reset()
x = var('x')
y = function('y')(x)
eqd = diff(y, x) == 2*x*(1 + y^2)
show(desolve(eqd, y))
sol = desolve(eqd, y)
_C = var('_C')
[sol1] = solve(sol, y(x))
show(sol1)
f(x, _C) = sol1.rhs()
plot(f(x, 0), x, -10, 10)
```

```python
#1b
reset()
x = var('x')
y = function('y')(x)
eqd = (x^2 - 1)*diff(y, x) + 2*x*y^2 == 0
show(desolve(eqd, y))
sol = desolve(eqd, y)
_C = var('_C')
[sol1] = solve(sol, y(x))
show(sol1)
f(x, _C) = sol1.rhs()
plot(f(x, 0), x, 1, 20, ymin=-5, ymax=5)
```

```python
#1c
reset()
x = var('x')
y = function('y')(x)
eqd = 2*x^2*diff(y, x) == x^2 + y^2
sol = desolve(eqd, y)
show(sol)
yy, _C= var('yy, _C')
f(x, yy, _C) = sol.substitute(y(x) == yy)
implicit_plot(f(x, yy, 1), (x, -20, 20), (yy, -20, 20))
```

```python
#3b
reset()
x, k = var('x, k')
k = 1
y = function('y')(x)
eqd = diff(y, x) + (k/x) * y == x^3
sol(x, _C) = desolve(eqd, y, ivar = x)
show(sol(x, _C))
_C = var('_C')
list_sol = [sol(x, 10000000000*i) for i in [10..20]]
plot(list_sol, x, 1, 10)
```
