```python
# 8a
reset()
t,s=var('t,s')
f(t,s)=1-(s*cos(4*t)*cos(t))/sqrt(1-(s^2)*(cos(4*t)^22)*(sin(t)^2))
x(t,s)=f(t-pi/2,s)
y(t,s)=f(t,s)
s=0.5
parametric_plot([x(t,s), y(t,s)], (t, 0, 2*pi))
```

```python
# 8b
g = parametric_plot((x(t,1/10), y(t, 1/10)), (t, -10, 10))
for k in [2..10]:
    g1= parametric_plot([x(t,k/10), y(t, k/10)], (t, -10, 10))
    g = g + g1
show(g)
```
