```python
#1a
reset()
t,w0 = var('t, w0')
x = function('x')(t)
eqd = diff(diff(x,t), t) + w0^2 * x == 0
show(eqd)
assume(w0>0)
sol = desolve(eqd, [x,t])
show(sol)
```

```python
#1b
x0, v0 = var('x0, v0')
f(t, w0, x0, v0) = desolve(eqd, [x, t], [0, x0, v0])
show(f(t, w0, x0, v0))
```

```python
#1c
L = 39.24/100 #transformam in metri
g = 9.81
w0 = sqrt(g/L)
x0 = 15/100 #transformam in metri
v0 = 0 
R = sqrt(x0^2 + (v0/w0)^2) #R - amplitudinea
d =  arctan(v0/(x0 * w0))#d - faza miscarii
T = (2*pi) / w0 # T - perioada
x0 = 15/100 #transformam in metri

z = function('x')(t)
z(t) = R*cos(w0 * t -d)
show(z(t).full_simplify())
plot(z(t), t, -T, T)
```

```python
#2a
reset()
w0, l, t = var('w0, l, t')
x = function('x')(t)
eqd = diff(x, t, 2) + l * diff(x, t) + w0^2 * x == 0
show(eqd)
assume(l^2 > 4 * w0^2)
sol = desolve(eqd, [x, t])
show(sol)
```

```python
#2b
l, w0, x0, v0 = var('l, w0, x0, v0')
assume(l^2 > 4 * w0^2)
f(t, l, w0, x0, v0) = desolve(eqd, [x, t], [0, x0, v0])
show(f(t, 25, 10, 1, 5))
plot(f(t, 25, 10, 1, 5), x, 0, 1)
```

```python
#2d
l, x0, v0 = var('l, x0, v0')
assume(l > 0)
f(t, l, x0, v0) = desolve(eqd, [x, t], [0, x0, v0])
show(f(t, l, x0, v0))
show(f(t, 20, 1, 5))
plot(f(t, 20, 1, 5), xmin = 0, xmax = 2)
```

```python
#3a
reset()
w0, w, F0, t = var('w0, w, F0, t')
x = function('x')(t)
eqd = diff(x, t, 2) + w0^2 * x == F0 * cos(w *t)
show(eqd)
assume(w0 > 0)
assume(w0 != w)
sol = desolve(eqd, [x, t])
show(sol)
```

```python
#3b
w, x0, v0 = var('w, x0, v0')
f(t, w0, w, F0, x0, v0) = desolve(eqd, [x, t], [0, x0, v0])
show(f(t, w0, w, F0, x0, v0))
show(f(t, 5, 5.5, 2, 0, 0))
plot(f(t, 5, 5.5, 2, 0, 0), xmin = -5, xmax = 5)
```

```python
#3c
reset()
w0, F0, t = var('w0, F0, t')
x = function('x')(t)
eqd = diff(x, t, 2) + w0^2 * x == F0 * cos(w0 *t)
show(eqd)
assume(w0 > 0)
sol = desolve(eqd, [x, t])
show(sol)
```

```python
#3d
x0, v0 = var('x0, v0')
f(t, w0, F0, x0, v0) = desolve(eqd, [x, t], [0, x0, v0])
show(f(t, w0, F0, x0, v0))
show(f(t, 5, 2, 0, 0))
plot(f(t, 5, 2, 0, 0), xmin = -5, xmax = 5)
```
