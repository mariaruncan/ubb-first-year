```python
#2 a
reset()
s = var('s')
f = function('f')(s)
f(s) = s * (s + 2) * (1 - 2)
solutii = solve(f(s) == 0, s)
show(solutii)
```

```python
x1 = solutii[0].rhs()
diff(f, s)(x1)
# este mai mare ca 0, deci x1 = -2 instabil
```

```python
x2 = solutii[1].rhs()
diff(f, s)(x2)
# este mai mic decat 0, deci x2 = 0 local asimptotic stabil
```

```python
#2 b
t = var('t')
x = function('x')(t)
eqd = diff(x, t) == x * (x + 2) * (1 - x)
g1 = desolve_rk4(eqd, x, ics = [0, -3], output = 'plot', xmin = -2, xmax = 2, color = 'black')
g2 = desolve_rk4(eqd, x, ics = [0, -2], output = 'plot', xmin = -2, xmax = 2, color = 'blue')
g3 = desolve_rk4(eqd, x, ics = [0, -1], output = 'plot', xmin = -2, xmax = 2, color = 'green')
g4 = desolve_rk4(eqd, x, ics = [0, 0], output = 'plot', xmin = -2, xmax = 2, color = 'red')
g5 = desolve_rk4(eqd, x, ics = [0, 0.5], output = 'plot', xmin = -2, xmax = 2, color = 'yellow')
g6 = desolve_rk4(eqd, x, ics = [0, 1], output = 'plot', xmin = -2, xmax = 2, color = 'cyan')
g7 = desolve_rk4(eqd, x, ics = [0, 2], output = 'plot', xmin = -2, xmax = 2, color = 'purple')
show(g1+g2+g3+g4+g5+g6+g7)
```

```python
#3 a
reset()
t = var('t')
x = function('x')(t)
y = function('y')(t)
eq1 = diff(x, t) == 3 * x + y
eq2 = diff(y, t) == x + 3 * y
syst = [eq1, eq2]
C1, C2 = var('C1, C2')
sol = desolve_system(syst, [x, y], ics = [0, C1, C2])
show(sol)
```

```python
#3 b
sol2 = desolve_system(syst, [x, y], ics = [0, 3, 3])
show(sol2)
```
